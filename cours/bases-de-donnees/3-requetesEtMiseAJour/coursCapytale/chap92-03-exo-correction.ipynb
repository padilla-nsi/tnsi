## Donner le code `SQL` de chacune des requêtes ci-dessous.

#### 1. Le titre des livres empruntés.

```sql
SELECT livre.titre FROM emprunt
JOIN livre ON emprunt.isbn = livre.isbn;
```


#### 2. Le titre des livres empruntés à rendre avant le 31/03/2020.

```sql
SELECT titre FROM emprunt
JOIN livre ON emprunt.isbn = livre.isbn
WHERE emprunt.retour <= '2020-03-31';
```


#### 3. Le nom et prenom de l'auteur du livre ' 1984' .

```sql
SELECT auteur.nom, auteur.prenom FROM livre
JOIN auteur_de ON livre.isbn = auteur_de.isbn
JOIN auteur ON auteur.a_id = auteur_de.a_id
WHERE titre = '1984';
```


#### 4. Le nom et le prenom des usagers ayant emprunté des livres, sans doublons.
(i. e. si un usager a emprunté plusieurs livres, il ne doit apparaître qu'une fois dans le résultat).

```sql
SELECT DISTINCT usager.nom, usager.prenom FROM emprunt
JOIN usager ON emprunt.code_barre = usager.code_barre;
```


#### 5. Même requête que précédemment, avec les noms triés par ordre alphabétique.

```sql
SELECT DISTINCT nom, prenom FROM emprunt
JOIN usager ON emprunt.code_barre = usager.code_barre
ORDER BY nom, prenom ASC;
```


#### 6. Les titre des livres publiés strictement avant ' Dune' .

```sql
SELECT titre FROM livre
WHERE annee < (SELECT annee FROM livre 
                            WHERE titre = 'Dune');
```


#### 7. Les noms et prenoms des auteurs des livres trouvés à la question précédente.

```sql
SELECT auteur.nom, auteur.prenom FROM livre
JOIN auteur_de ON livre.isbn = auteur_de.isbn
JOIN auteur ON auteur.a_id = auteur_de.a_id
WHERE annee < (SELECT annee FROM livre 
                            WHERE titre = 'Dune')
ORDER BY nom;
```


#### 8. Comme la question précédente, en retirant les doublons.

```sql
SELECT DISTINCT auteur.nom, auteur.prenom FROM livre
JOIN auteur_de ON livre.isbn = auteur_de.isbn
JOIN auteur ON auteur.a_id = auteur_de.a_id
WHERE annee < (SELECT annee FROM livre 
                            WHERE titre = 'Dune')
ORDER BY nom;
```


#### 9. Le nombre de résultats trouvés à la question précédente.

```sql
SELECT COUNT(DISTINCT auteur.nom) FROM livre
JOIN auteur_de ON livre.isbn = auteur_de.isbn
JOIN auteur ON auteur.a_id = auteur_de.a_id
WHERE annee < (SELECT annee FROM livre 
                            WHERE titre = 'Dune');
```
