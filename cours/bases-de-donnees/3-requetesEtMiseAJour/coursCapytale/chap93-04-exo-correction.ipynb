## Exercice 1

Identifier le nom du ou des professeurs enseignant la spécialité pour le groupe dont l'identifiant `grpid` est `1SVTG1`.  
    
</div>

````sql
SELECT professeur.nom FROM enseigne
JOIN professeur ON professeur.pid = enseigne.pid
WHERE enseigne.grpid = '1SVTG1';
````

## Exercice 2
    
a. Afficher les `nom`, `prenom` et `grpid` de tous les lycéens. Pour cela, utiliser une jointure entre les tables `lyceen` et `affectation`.
    
Exemple de résultat attendu :

## Exercice 3 - jointure double
    
Afficher les noms des élèves suivant la spécialité d'intitulé "Physique-Chimie".
    
</div>

````sql
SELECT lyceen.nom FROM groupe
JOIN affectation ON affectation.grpid = groupe.grpid
JOIN lyceen ON lyceen.id = affectation.id
WHERE intitule = 'Physique-Chimie';
````

## Exercice 4
    
a. Utiliser une jointure entre les tables ```Lyceen```, ```Affectation``` et ```Groupe``` pour afficher pour chaque lycéen de Terminale ses `nom` et `prenom` avec l'`intitulé` de chacune des spécialités où il est inscrit.
